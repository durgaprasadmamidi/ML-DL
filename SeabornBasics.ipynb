```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('mall_customers.csv')
```

```python
data.head()
```

```python
data.isnull().any()
#data.isnull().any().any()
```

```python
plt.rcParams['figure.figsize'] = (18,5) #width, height in inches
plt.subplot(1,2,1)
sns.set(style='whitegrid')
sns.distplot(data.iloc[:,3:4])
plt.subplot(1,2,2)
sns.distplot(data['Age'],color='red')
plt.title('Distribution of age', fontsize = 30)
plt.xlabel('Range of Age')
plt.ylabel('Count')
plt.show()
```

```python
data['Genre'].value_counts()
```

```python
labels=['Frmale','Male']
size=data['Genre'].value_counts()
colors=['lightgreen','orange']
explode=[0,0.05] # elplodes the second slice i.e male
plt.rcParams['figure.figsize']= (5,5)
plt.pie(size,labels=labels,colors=colors,explode=explode,autopct='%0.2f%%')
plt.legend()
```
