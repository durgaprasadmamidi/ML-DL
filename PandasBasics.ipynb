```python
employes = pd.DataFrame({"Empcode":[1,2,3,4],"Name":["Durga","Abjit","Elle","Austin"],"Occupation":["ML Engineer","AI Engineer","ML Engineer","App Developer"],"Age":[22,21,21,28]})
employes
```

```python
employes.columns
```

###### assigning columns 

```python
employes.columns=['Empcode', 'EmpName', 'EmpOccupation', 'EmpAge']
employes
```

###### rename perticular column

```python
employes.rename(columns={"Empcode":"EmpId"})
```

```python
employes
```

###### witch the help of inplace we can make changes to current object

```python
employes.rename(columns={"Empcode":"EmpId"},inplace=True)
```

```python
employes
```

```python
employes.drop("EmpId",axis=1)
```

```python
employes
```

```python
employes.drop(1,axis=0)
```

###### removing multiple columns using index numbers

```python
employes.drop(employes.columns[[0,3]],axis=1)
```

###### add a column to dataframe

```python
employes['Empcity']=["Vizag","Hyderabad","Vizag","Delhi"]
employes
```

###### inserting a column into dataframe

```python
employes.insert(loc=1,column="Surname",value=["M","K","P","E"])
employes
```

###### adding a new row

```python
new={'EmpId':5, 'Surname':"N", 'EmpName':"Noel", 'EmpOccupation':"UI/UX", 'EmpAge':25, 'Empcity':"Mumbai"}
employes.append(new,ignore_index=True)
```

```python
employes
```

```python
new={'EmpId':5, 'Surname':"N", 'EmpName':"Noel", 'EmpOccupation':"UI/UX", 'EmpAge':25, 'Empcity':"Mumbai"}
employes=employes.append(new,ignore_index=True)
```
